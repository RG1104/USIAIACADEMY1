```python
# defining different iteration count function with an assumption of 1 second per iteration,
def iterator_count(n):
    count = 0
    for i in range(n):
        count += 1
    return count

def iterator_count2(n):
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count

def iterator_count3(n):
    count = 0
    for i in range(n):
        for j in range(i):
            count += 1
    return count

def iterator_count4(n):
    count_i = 0
    count_j = 0
    for i in range(n):
        for j in range(i):
            count_j += 1
        count_i += 1
    return count_i + count_j
```

```python
List =[i for i in range(100)] #generating dataset for plotting with list comprehension.

Operator_1 = []
Operator_2 = []
Operator_3 = []
Operator_4 = []

for i in List:
    t1 = iterator_count(i)
    Operator_1.append(t1)
    t2 = iterator_count2(i)
    Operator_2.append(t2)
    t3 = iterator_count3(i)
    Operator_3.append(t3)
    t4 = iterator_count4(i)
    Operator_4.append(t3)
    
```

```python
plt.figure(figsize=(10,8))
plt.plot(Operator_1, label = 'Operator_1')
plt.plot(Operator_2, label = 'Operator_2')
plt.plot(Operator_3, label = 'Operator_3')
plt.plot(Operator_4, label = 'Operator_4')
plt.xlabel ('Number of Iterations')
plt.ylabel ('time in seconds')
plt.legend()
plt.show()

```
